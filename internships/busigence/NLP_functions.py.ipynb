```python
#opening files
amazon_file = open(r"C:\Users\Mohit\Desktop\Language_1\Round1_Problem1-of-2_Dataset_amazon_cells_labelled.txt")
imdb_file = open(r"C:\Users\Mohit\Desktop\Language_1\Round1_Problem1-of-2_Dataset_imdb_labelled.txt")
```

```python
#reading both files in pandas
import pandas as pd
amazon_csv = pd.read_csv(r"C:\Users\Mohit\Desktop\Language_1\Round1_Problem1-of-2_Dataset_amazon_cells_labelled.txt", delimiter = "\t", names = ("Text", "Sentiment"))
imdb_csv = pd.read_csv(r"C:\Users\Mohit\Desktop\Language_1\Round1_Problem1-of-2_Dataset_imdb_labelled.txt", delimiter = "\t", names = ("Text", "Sentiment"))
```

```python
#adding the two csv's
data = amazon_csv.append(imdb_csv, ignore_index = True)
```

```python
#printing the original data
print(data)
```

```python
#Removing all HTML characters in the data
import re
data_1 = []
for i in range(0, 1748):
    a = re.sub(r'[^ a-zA-z0-9,.;:!]', ' ', data['Text'][i])
    data_1.append(a)
data_1 = pd.DataFrame(data_1, columns = ['Text'])
data_1 = data_1.join(data['Sentiment'])
print(data_1)
```

```python
#Removing website links from the data
data_2 = []
for i in range(0, 1748):
    b = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", data_1['Text'][i])
    data_2.append(b)
data_2 = pd.DataFrame(data_2, columns = ['Text'])
data_2 = data_2.join(data['Sentiment'])
print(data_2)
```

```python
#Demojizing the strings
import emoji
data_3 = []
for i in range(0, 1748):
    c = emoji.demojize(data_2['Text'][i])
    data_3.append(c)
data_3 = pd.DataFrame(data_3, columns = ['Text'])
data_3 = data_3.join(data['Sentiment'])
print(data_3)
```

```python
#Removing all punctuations except fullstop (.), exclamation (!) and comma (,)
data_4 = []
for i in range(0, 1748):
    d = re.sub(r'[^ a-zA-z0-9,.!]', ' ', data_3['Text'][i])
    data_4.append(d)
data_4 = pd.DataFrame(data_4, columns = ['Text'])
data_4 = data_4.join(data['Sentiment'])
print(data_4)
```

```python
#Removing major stopwords
import nltk
from nltk.corpus import stopwords
data_5 = []
for i in range(0, 1748):
    rev_amazon = data_4['Text'][i].split()
    rev_amazon = [word for word in rev_amazon if not word in set(stopwords.words('english'))]
    data_5.append(rev_amazon)
    data_5[i] = ' '.join(data_5[i])
data_5 = pd.DataFrame(data_5, columns = ['Text'])
data_5 = data_5.join(data['Sentiment'])
print(data_5)
```

```python
#Lemmatizing all the words
for i in range(0, 1000):
    data_5['Text'][i] = data_5['Text'][i].lower()

from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
data_6 = []
for i in range(0, 1748):
    rev_amazon = data_5['Text'][i].split()
    rev_amazon = [lemma.lemmatize(word) for word in rev_amazon]
    data_6.append(rev_amazon)
    data_6[i] = ' '.join(data_6[i])
data_6 = pd.DataFrame(data_6, columns = ['Text'])
data_6 = data_6.join(data['Sentiment'])
print(data_6)
```

```python
#Tokenizing the data
from nltk.tokenize import word_tokenize
data_7 = []
for i in range(0, 1000):
    g = word_tokenize(data_6['Text'][i])
    data_7.append(g)
print(data_7)
```

```python
#Conversion of data into vector form
import gensim
from nltk.corpus import abc

model = gensim.models.Word2Vec(data_7, size = 20)
model.train(data_7, total_examples = 1000, epochs=10)
words = list(model.wv.vocab)
data_8 = model[model.wv.vocab]
print(data_8)
data_8 = pd.DataFrame(data_8)
data_8.to_csv('output.csv', index = False)
print(data_8)
```
