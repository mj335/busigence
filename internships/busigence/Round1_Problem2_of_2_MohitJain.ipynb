# Slot filling seq to seq model

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
#from keras.utils.vis_utils import plot_model
# Define an input sequence and process it.
latent_dim = 256
encoder_inputs = Input(shape=(None,))
xs = Embedding(2500, latent_dim, input_length = x.shape[1])(encoder_inputs)
xs, state_h, state_c = LSTM(latent_dim,
                           return_state=True)(xs)
encoder_states = [state_h, state_c]

# Set up the decoder, using `encoder_states` as initial state.
decoder_inputs = Input(shape=(None,))
xs = Embedding(2500, latent_dim, input_length = y.shape[1])(decoder_inputs)
xs = LSTM(latent_dim, return_sequences=True)(xs, initial_state=encoder_states)
decoder_outputs = Dense(y_en.shape[2], activation='softmax')(xs)

# Define the model that will turn
# `encoder_input_data` & `decoder_input_data` into `decoder_target_data`
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile & run training
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])
# Note that `decoder_target_data` needs to be one-hot encoded,
# rather than sequences of integers like `decoder_input_data`!
model.fit([x, y], y_en,
          batch_size=256,
          epochs=15,
          validation_split=0.2)
```
